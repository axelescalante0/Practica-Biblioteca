```python
#Libreria y Ruta de accespo
import sys
import os

sys.path.append(os.path.abspath('C:/Users/Axel/Desktop/Practica-Biblioteca'))

import pandas as pd
```

```python
programas = pd.read_excel('C:/Users/Axel/Desktop/Practica-Biblioteca/data/procesados/bibliografia_programas_proce.xlsx')
```

```python
programas.head()
```

```python
#pasar a minusculas titulo y autor
programas['titulo'] = programas['titulo'].str.lower()
programas['autor'] = programas['autor'].str.lower()

#sacar espacios en blanco al principio y al final
programas['titulo'] = programas['titulo'].str.strip()
programas['autor'] = programas['autor'].str.strip()

#sacar los puntos al final del titulo y autor
programas['titulo'] = programas['titulo'].str.rstrip('.')
programas['autor'] = programas['autor'].str.rstrip('.')

```

```python
#guardar el archivo en la carpeta de procesados
programas.to_excel('C:/Users/Axel/Desktop/Practica-Biblioteca/data/procesados/bibliografia_programas_proce.xlsx', index=False)
```
