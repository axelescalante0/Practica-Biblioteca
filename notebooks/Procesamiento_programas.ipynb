```python
#Libreria y Ruta de accespo
import sys
import os

sys.path.append(os.path.abspath('C:/Users/Axel/Desktop/Practica-Biblioteca'))

import pandas as pd
```

```python
programas = pd.read_excel('C:/Users/Axel/Desktop/Practica-Biblioteca/data/procesados/cp_bibliografia_programas.xlsx')
```

```python
programas.head()
```

```python
#separar en la columna de bibliografía titulo y autor: usar separador '$' donde el primer elemento es el título y el segundo el autor generando dos columnas nuevas
programas[['titulo','autor']] = programas['Bibliografía'].str.split('$', expand=True)
programas.drop(columns=['Bibliografía'], inplace=True)

#en la columna asigntura separar el nombre de la asignatura, es el primer elemento antes del guión '-' lo demas se elimina
programas['Asignatura'] = programas['Asignatura'].str.split('-').str[0] 

```

```python
#guardar el archivo en la carpeta de procesados
programas.to_excel('C:/Users/Axel/Desktop/Practica-Biblioteca/data/procesados/bibliografia_programas_proce.xlsx', index=False)
```
